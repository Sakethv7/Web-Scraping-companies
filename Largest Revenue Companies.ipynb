```python
from bs4 import BeautifulSoup

# Assuming 'page.text' contains the HTML content of the page
soup = BeautifulSoup(page.text, 'html.parser')

# Find the target table
target_table = soup.find('table', {'class': 'wikitable sortable'})

# Extract all the rows from the table
rows = target_table.find_all('tr')

# Initialize an empty list to store the extracted row data
data_rows = []

# Extract data from each row
for row in rows:
    # Extract both header and data cells
    columns = row.find_all(['th', 'td'])  
    data = [col.get_text(strip=True) for col in columns]
    
    # Print data to debug and check the extracted columns
    print(f"Extracted data: {data}")
    
    # Append data to data_rows
    data_rows.append(data)

# Now, let's print data_rows to see what we have before putting it into a DataFrame
print(data_rows)
```

```python
# Remove the 'Revenue growth' header if it's not needed in the data
if 'Revenue growth' in world_tables_titles:
    world_tables_titles.remove('Revenue growth')
```

```python
headers = ['Rank', 'Name', 'Industry', 'Revenue(USD millions)', 'Revenue growth', 'Employees', 'Headquarters']
import pandas as pd

# Create a DataFrame using the extracted data
df_data = pd.DataFrame(data_rows[1:], columns=headers)  # Skip the first row if it's headers

# Display the DataFrame
print(df_data)

```
